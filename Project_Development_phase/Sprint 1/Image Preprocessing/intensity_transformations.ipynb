```python
from skimage.io import imread
import matplotlib.pyplot as plt
import numpy as np
```

```python
live = imread('/content/drive/MyDrive/dataset/fruit-dataset/train/Apple___healthy/0181db4e-718e-41bc-b6ed-92863471fa85___RS_HL 5941.JPG')
mask = imread('/content/drive/MyDrive/dataset/veg-dataset/train-set/Tomato___Septoria_leaf_spot/026c9801-8f0c-41b6-adfe-12704eafa73d___Keller.St_CG 1939.JPG')

plt.figure(figsize=(10, 10))
plt.subplot(121), plt.imshow(live, cmap='gray')
plt.subplot(122), plt.imshow(mask, cmap='gray')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(121), plt.imshow(live, cmap='gray')
plt.subplot(122), plt.imshow(live - 20, cmap='gray')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(131), plt.imshow(mask - live, cmap='gray')
plt.subplot(132), plt.imshow(-(mask - live + 128), cmap='gray')
plt.subplot(133), plt.imshow(mask - live + 128, cmap='gray')
plt.show()
```
